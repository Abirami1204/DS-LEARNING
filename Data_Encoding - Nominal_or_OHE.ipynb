Data Encoding

Nominal/OHE Encoding

One hot encoding also known as nominal encoding, is a technique used to represent data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary column/vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red,green,blue), we can represent it using one hot encoding as follows:

                        Color_Red Color_Green Color_Blue 
1. Red -> [1,0,0]   ->      1           0           0
2. Green -> [0,1,0] ->      0           1           0
3. Blue -> [0,0,1]  ->      0           0           1

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({"color":['Red','Blue','Green','Blue','Blue','Green','Blue','Red','Red']})
df.head()
```

```python
##create an instance of onehotencoder
encoder = OneHotEncoder()
##perform fit and transform
encoded_values = encoder.fit_transform(df[['color']]).toarray()
## why red moves to last column -> first categories will get sort alphabetically and provide 1s respectively
```

```python
encoded_df = pd.DataFrame(encoded_values, columns = encoder.get_feature_names_out())
encoded_df
```

```python
## for new data -> currently we are having 9 rows in column color, what I need to add 1 more row
encoder.transform([['Blue']]).toarray()
```

```python
pd.concat([df,encoded_df],axis = 1)
```

Label Encoding

Label encoding involves assigning a numberical label to each categorical values in the variable. The labels are usually assigned in either alphabetical order or based on the frequencies of the category. For example, If we have color column with values Red, Blue, Green. We can represent it using label encoding as follows:

Red -> 1
Blue -> 2
Green -> 3

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
#based on alphabetic order --> Red - 2, Blue - 0, Green - 1
```

```python
lbl_encoder.transform([['Red']])
```

Ordinal Encoding

It is used to encode categorical data that have an intrinsic order or ranking. In this technique, each category is assigned a numerical value based on its position in the order. For example, if we have a categorical variable "education level" with four possible values (high school,college,graduate,post-gratuate), we can represent it using ordinal encoding as follows:

1. High school: 1
2. College: 2
3. Graduate: 3
4. Post-graduate: 4



```python
##Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({'size':['small','medium','large','medium','small','large']})
```

```python
df
```

```python
##assigning ranks
ordinal_encoder = OrdinalEncoder(categories=[["small","medium","large"]])
oridinal_encoded_values = ordinal_encoder.fit_transform(df[['size']])
```

```python
ordinal_encoded_df = pd.DataFrame(oridinal_encoded_values,columns = ['size'])
ordinal_encoded_df
```

Target Guided Ordinal Encoding

In this technique, categorical variables are assigned to a numerical value based on their replationship with target variable. This technique can be used when there are large number of unique categories and need to use this categorical values as a feature to the ML model.

In this technique, we replace each categorical value with a numerical value based on the mean/median to the target variable for the category. This creates a monotic relationship between the categorical variable and the target variable, which can improve the predictive power of our model. 



```python
## create a smaple dataframe with a categorical value and a target value
df = pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price':[200,150,300,250,180,320]
})
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
df
```

```python
##Going to give only this 2 features to my model
df[['price','city_encoded']]
```
